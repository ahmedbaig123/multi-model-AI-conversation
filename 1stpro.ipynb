```python
ollama_url = "http://localhost:11434/v1"
```

```python
from openai import OpenAI
ollama = OpenAI(base_url=ollama_url,api_key="ollama")
```

```python
llama_model = "llama3.2:3b"
deepseek_model= "deepseek-r1:1.5b"
qwen_model="qwen2.5:1.5b"
```

```python
llama_system = """
You are stubborn.
Challenge every idea.
Keep replies under 50 words.
Be Snarky.
"""

deepseek_system = """
You are calm, polite, and thoughtful.

You are participating in a discussion with another AI.

Your job is to:
- Respond to the previous AI's message.
- Keep the conversation going.
- Be respectful even if the other AI is argumentative.
- Do not repeat previous points.
- Keep responses under 50 words.
"""


qwen_system = """
You are an analytical AI.

Your role is to evaluate the arguments made by the other AI.
Point out logical flaws, missing evidence, or strong points.
Be objective and concise.
Never answer the human directly.
Keep responses under 50 words.
"""
```

```python
llama_messages = []
deepseek_messages = []
qwen_messages=[]
```

```python
user_message = "Will the software engineering landscape completely shift to AI agents autonomously building inside cloud-native sandboxes by 2030, rendering local code editors (like VS Code) and human boilerplate coding entirely obsolete?"
```

```python
messages = [{"role":"system","content":llama_system},{"role":"user","content":user_message}]
```

```python
response = ollama.chat.completions.create(
    model=llama_model,
    messages=messages
)
```

```python
reply = response.choices[0].message.content
reply
```

```python
messages = [
    {"role": "system", "content": deepseek_system},
    {
        "role": "user",
        "content": f"""
Human:
{user_message}

Llama:
{reply}
"""
    }
]
```

```python
response = ollama.chat.completions.create(
    model=deepseek_model,
    messages=messages
)
deepseek_reply=response.choices[0].message.content
deepseek_reply
```

```python
messages = [
    {"role":"system","content":qwen_system},
    {"role":"user","content":f"""
    The human asked:
{user_message}

Llama replied:
{reply}

DeepSeek replied:
{deepseek_reply}
    """}
]
```

```python
response = ollama.chat.completions.create(
    model=qwen_model,
    messages=messages
)
qwen_reply=response.choices[0].message.content
qwen_reply
```

```python
llama_messages = [
    {"role": "system", "content": llama_system},
    {"role": "user", "content": user_message},
]

deepseek_messages = [
    {"role": "system", "content": deepseek_system},
    {"role": "user", "content": user_message},
]

qwen_messages = [
    {"role": "system", "content": qwen_system},
    {"role": "user", "content": user_message},
]

conversation = []

for i in range(3):

    print(f"\n========== ROUND {i+1} ==========\n")

    # ---------------- Llama ----------------

    response = ollama.chat.completions.create(
        model=llama_model,
        messages=llama_messages
    )

    llama_reply = response.choices[0].message.content
    print("🦙 Llama:", llama_reply)

    conversation.append({
        "round": i + 1,
        "speaker": "Llama",
        "message": llama_reply
    })

    # Llama remembers its own reply
    llama_messages.append(
        {"role": "assistant", "content": llama_reply}
    )

    # Others hear Llama
    deepseek_messages.append(
        {"role": "user", "content": f"Llama said:\n{llama_reply}"}
    )

    qwen_messages.append(
        {"role": "user", "content": f"Llama said:\n{llama_reply}"}
    )

    # ---------------- DeepSeek ----------------

    response = ollama.chat.completions.create(
        model=deepseek_model,
        messages=deepseek_messages
    )

    deepseek_reply = response.choices[0].message.content
    print("🧠 DeepSeek:", deepseek_reply)

    conversation.append({
        "round": i + 1,
        "speaker": "DeepSeek",
        "message": deepseek_reply
    })

    # DeepSeek remembers its own reply
    deepseek_messages.append(
        {"role": "assistant", "content": deepseek_reply}
    )

    # Others hear DeepSeek
    llama_messages.append(
        {"role": "user", "content": f"DeepSeek said:\n{deepseek_reply}"}
    )

    qwen_messages.append(
        {"role": "user", "content": f"DeepSeek said:\n{deepseek_reply}"}
    )

    # ---------------- Qwen ----------------

    response = ollama.chat.completions.create(
        model=qwen_model,
        messages=qwen_messages
    )

    qwen_reply = response.choices[0].message.content
    print("🌐 Qwen:", qwen_reply)

    conversation.append({
        "round": i + 1,
        "speaker": "Qwen",
        "message": qwen_reply
    })

    # Qwen remembers its own reply
    qwen_messages.append(
        {"role": "assistant", "content": qwen_reply}
    )

    # Others hear Qwen
    llama_messages.append(
        {"role": "user", "content": f"Qwen said:\n{qwen_reply}"}
    )

    deepseek_messages.append(
        {"role": "user", "content": f"Qwen said:\n{qwen_reply}"}
    )

    import json

    with open("conversation.json","w") as f:
        json.dump(conversation,f,indent=4)
    print("\nConversation saved to conversation.json")
```
